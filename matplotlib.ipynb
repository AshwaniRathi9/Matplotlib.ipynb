```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,'r*')
```

```python
plt.plot(x,y,'bo',linestyle= 'dashed')
```

```python
# createing subplot
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(y,x,'r*')

plt.subplot(2,2,4)
plt.plot(y,x,'g*')

plt.subplot(2,2,3)
plt.plot(x,y,'b+')

plt.subplot(2,2,1)
plt.plot(y,x,'r--')
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data)
```

```python
# data to plot
labels = 'python', 'C++', 'Java', 'Ruby'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
